### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

# Generate x and y values
x = list(range(0, 11))           # x values from 0 to 10
y = [i**2 for i in x]            # y values as square of x

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add title and labels
plt.title('Square of Numbers from 0 to 10')
plt.xlabel('X values')
plt.ylabel('Y = X^2')

# Add grid
plt.grid(True)

# Display the plot
```
