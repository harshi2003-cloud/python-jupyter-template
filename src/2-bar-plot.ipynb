### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'yellow']  # Different color for each bar

# Create bar plot
plt.bar(categories, values, color=colors)

# Add title and axis labels
plt.title('Bar Plot of Category Values')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```
